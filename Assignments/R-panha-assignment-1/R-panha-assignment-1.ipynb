# 1. Calculate the following

## a. Σ with and without loop.

```r
# Σ from a=5 to a=100 of (a^2 + 5a^3) with loop
sum <- 0

for (a in 5:100)
    {
    sum <- sum + (a^4+5*a^3)
}

sum
```

```r
# Σ from a=5 to a=100 of (a^2 + 5a^3) without loop
a = 5:100
sum(a^4+5*a^3)
```

## b. Calculate U<sub>40</sub>

```r
n <- 40
u <- 1:2
for (i in 3:40){
    u <- append(u, u[i-1] + u[i-2])
}
u[n]
```

# 2. Solve the following system of linear equations using Gaussian elimination (A x =y)

x + 2y + 3x = 9

2x - y + 2z = −3

3x + 3y + z = 5

```r
# required libraries
install.packages("rgl", repos="http://R-Forge.R-project.org")
install.packages("matlib")
```

```r
library(matlib)
A <- matrix(c(1, 2, 3,
              2, -1, 2,
              3,  3, 1), 3, 3, byrow=TRUE)
o <- c(9, -3, 5)
res <- gaussianElimination(A, o)
res
paste("Result : x =",res[10],' y = ', res[11], ' z =',res[12])
```

# 4. COVID-19 Dataset in Japan

## a. Data Source detail

```r
library(tidyverse)
dataset <- read_csv("./covid_jpn_prefecture.csv")
covid_JPN_dataset <- dataset
covid_JPN_dataset
```

## b. Explain the Unit & Necessity of each variable  

```r
names(covid_JPN_dataset)
```

### This dataset has 8 variables:
 1. Date: is the observed date of the cases on the specified date, to be able to see the timestamp for the total cases
 2. Prefecture: provinces and cities
 3. Positive: positive cases of COVID-19 in Japan
 4. Tested: number of population tested in each prefecture
 5. Discharged: number of cured patients from COVID-19
 6. Fatal: Number of deaths from COVID-19
 7. Hosp_require: number of patients that required hospital treatments
 8. Hosp_severe: number of severe cases in each prefecture

## c. Find the missing values(rows & columns) and replace them with mean(Tidy Dataset)

```r
covid_JPN_dataset[] <- lapply(covid_JPN_dataset, function(z) { 
  z[is.na(z)] <- mean(z, na.rm = TRUE)
  z
})
covid_JPN_dataset
```

## d. Generate the two new variables(Var1:Mean, Var2: Median from available variable)

```r
covid_JPN_dataset %>% mutate(Mean = mean(Positive),Median = median(Positive))
```

## e. Rename the two existing variables

```r
covid_JPN_dataset %>% rename(Cases = Positive, Death = Fatal)
```

## f. Create a plot using following instructions (using 7 layers of Grammar of Graphics)

### i. Choose x and y axis(aes)

```r
#Choose Date as x axis and Cases as Y axis
dataJPcovid19 <- covid_JPN_dataset %>% count(Date, wt = Positive)

Date <- dataJPcovid19[['Date']]
Positive <- dataJPcovid19[['n']]

dataJPcovid19
```

## ii. geom_ point() - specify the parameters, size : 5, color: red, alpha: ⅕

```r
library(ggplot2)
gp = geom_point(alpha = 1/5, color = "red", size = 5)
gp

f <- ggplot(x = dataJPcovid19, mapping = aes(x = Date, y = Positive)) + gp
f
```

## ii. Use Facet grid, cartesian coordinates & geom_smooth()

```r
# facet_grid

f + facet_grid(Positive)
```

```r
# coord_cartesian

f + coord_cartesian(expand = FALSE)
```

```r
#geom_smooth

f + geom_smooth()
```

## iv. Assign the title to x, y and graph 

```r
f <- f + ggtitle("Number of COVID-19 cases in JAPAN") +
  xlab("Date") + ylab("Number of Positive cases")
f
```
