# Ex4: 4x4 Matrix arithmetic operations

```r
x <- matrix(1:16, nrow = 4, ncol = 4)
y <- matrix(17:32, nrow = 4, ncol = 4)


#(a). x + y
print('a')
a <- x + y
a

#(b). x - y 
print('b')
b <- x - y
b

#(c). x*y
print('c')
c <- x * y
c

#(d). x%*%y
print('d')
d <- x %*% y
d

#(e). Difference between (c) &(d) 
print('e')
e <- c - d
e

#(f). Inverse of y 
print('f')
#solve(y)
print("Inverse doesn't exist")


#(g). z<-x/y 
print('g')
z <- x / y
z



#(h). Transpose of z 
print('h')
h <- t(z)
h

#(i). Diagonal of transpose matrix 
print('i')
diagmatrix <- diag(h)
diagmatrix
```

# Ex10: ifelse statement
Syntax: if (condition) {...} else {...}. && and || can be used in the condition. If else statement can be nested.

```r
test <- c(1:10)
test 
print("Print out those sample numbers that are even using if else statement")
for (thistest in test)
{
 if (thistest %% 2 != 0) next
 else print(thistest)
}
```

# Ex11: BMI Calculator

```r
print("BMI Calc")
weight <- readline(prompt="Enter weight(kg): ")
weight <- as.numeric(weight)
height <- readline(prompt="Enter height(cm): ")
height <- as.numeric(height)
height <- height/100
BMI <- (weight/(height^2))
round(BMI, 1)

category <- ifelse(BMI >= 30, "obese", 
               ifelse(BMI >=25, "overweight", 
                  ifelse(BMI >=18.5, "normal", "underweight")))
category
```

# Ex12: List vs Data frames
- R list data type refers to an object consisting of an ordered collection of elements. The elements may be of different mode or type.
- R data.frame is a powerful data type, especially when processing table (.csv). It can store the data as row and columns according to the table. The difference between data frame and list is that the column data of data frame may be of different modes and attributes.




```r
df1 <- data.frame(animal = c("apple", "orange", "banana"),
color = c("red", "orange", "yellow"), amount = c(1, 2, 3))
df1

l1 <- list(animal = c("apple", "orange", "banana"),
color = c("red", "orange", "yellow"), amount = c(1, 2, 3))
l1
```

# Ex13: repeat vs while loop
repeat is similar to while and for loop, it will execute a block of commands repeatly till break. Also similiar to do while loop

```r
total <- 0
repeat { total <- total + 1; print(total); if (total > 6) break; }
total
```

while() loop will execute a block of commands until the condition is no longer satisfied.

```r
x <- 1
while(x < 5) {x <- x+1; if (x == 3) break; print(x); }
```
